```python
import matplotlib.pyplot as plt


y = sns.histplot(
      data=df,
      x = "Survived",
      hue = "Survived",
      bins = 2,
      palette = my_palette, 
  )
plt.xticks([0, 1], ["No (0)", "Yes(1)"])

y.bar_label(y.containers[0])
y.bar_label(y.containers[1])

plt.show()
```
